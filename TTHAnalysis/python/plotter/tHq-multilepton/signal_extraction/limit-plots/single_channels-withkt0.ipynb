## Limits for single channels

Just multiplying r limits for single channels with the cross sections to be quoted in the tables.

```python
import os
import pandas as pd
```

```python
# Hardcoded SM production cross sections for tHq, tHW, and ttH
xsec_thq_SM = 0.07096
xsec_thw_SM = 0.01561
xsec_tth_SM = 0.5071

# Hardcoded SM BRs for WW, ZZ, and tautau
SM_BR_ww = 0.214
SM_BR_zz = 0.026
SM_BR_tt = 0.063
SM_BR_bb = 0.583
SM_BR_gg = 0.00227

xsBR = ( xsec_thq_SM + xsec_thw_SM + xsec_tth_SM) * ( SM_BR_ww + SM_BR_zz + SM_BR_tt + SM_BR_bb + SM_BR_gg)
xsBR_tHonly = ( xsec_thq_SM + xsec_thw_SM) * ( SM_BR_ww + SM_BR_zz + SM_BR_tt + SM_BR_bb + SM_BR_gg)

print "Total cross section x BR (SM):        %5.2f pb" % xsBR
print "tH only cross section x BR (SM):      %6.3f pb" % xsBR_tHonly

# Cross sections for K6 and K7 for the inverted couplings:
df_xs_K6 = pd.read_csv('xsecs/xsecs_tH_ttH_WWZZttbbgg_K6.csv')
xsBRITCK6 = df_xs_K6.loc[df_xs_K6.cf==-1.0].loc[df_xs_K6.cv==1.0].tot
df_xs_K7 = pd.read_csv('xsecs/xsecs_tH_ttH_WWZZttbbgg_K7.csv')
xsBRITCK7 = df_xs_K7.loc[df_xs_K6.cf==-1.0].loc[df_xs_K6.cv==1.0].tot
df_xs_K5 = pd.read_csv('xsecs/xsecs_tH_ttH_WWZZttbbgg_K5.csv')
xsBRITCK5 = df_xs_K5.loc[df_xs_K5.cf==-1.0].loc[df_xs_K5.cv==1.0].tot
xsBRITCK5b = df_xs_K5.loc[df_xs_K5.cf==-1.0].loc[df_xs_K5.cv==1.0].thtot

xsBRKT0K5  = df_xs_K5.loc[df_xs_K5.cf==0.0].loc[df_xs_K5.cv==1.0].tot
xsBRKT0K5b = df_xs_K5.loc[df_xs_K5.cf==0.0].loc[df_xs_K5.cv==1.0].thtot

print "Total cross section x BR (ITC K6):    %5.2f pb" % xsBRITCK6
print "Total cross section x BR (ITC K7):    %5.2f pb" % xsBRITCK7
print "Total cross section x BR (ITC K5):    %5.2f pb" % xsBRITCK5
print "tH only cross section x BR (ITC K5b): %6.3f pb" % xsBRITCK5b
print "tH only cross section x BR (kt0 K5b): %6.3f pb" % xsBRKT0K5b

xsec_scale = {
    "SM" : xsBR,
    "SMtH" : xsBR_tHonly,
    "ITCK6" : float(xsBRITCK6),
    "ITCK7" : float(xsBRITCK7),
    "ITCK5" : float(xsBRITCK5),
    "ITCK5b" : float(xsBRITCK5b),
    "KT0K5b" : float(xsBRKT0K5b),
}
```

```python
#### Results (from output above):
def print_limits(row, tag=None, scale='SM'):
    tag = tag or os.path.basename(row.fname)
#     linepat = "%-40s %4.2f %4.2f: %5.2f  %5.2f +%-4.2f -%-4.2f pb   (r < %5.2f obs, r < %5.2f exp)"
#     linepat = "%-40s %4.2f %4.2f: $%5.2f$  &         $%5.2f\,^{+%5.2f}\,_{-%5.2f}$ pb   (r < %5.2f obs, r < %5.2f exp)"
    linepat = "%-40s %4.2f %4.2f: $%4.2f$ &         $%4.2f\,^{+%4.2f}_{-%4.2f}$ pb"
    print linepat % ( tag, row.cv, row.cf,
                      row.obs*xsec_scale[scale],
                      row.exp*xsec_scale[scale],
                     (row.onesigup-row.exp)*xsec_scale[scale],
                     (row.exp-row.onesigdown)*xsec_scale[scale], )
#                       row.obs, row.exp )
print "fname                                     cv   cf   obs(pb)  exp +expup -expdn (pb) (obs r<)    (exp r<)"
# for index,row in df_results.iterrows():
#     if not "K7" in row.fname: continue
#     if not "1_m1" in row.fname: continue
#     print_limits(row, scale='ITCK7')
    
# print "\n"
# for index,row in df_results.iterrows():
#     if not "K6" in row.fname: continue
#     if not "1_m1" in row.fname: continue
#     print_limits(row, scale='ITCK6')

# print "\n"
# for index,row in df_results.iterrows():
#     if not "K7" in row.fname: continue
#     if not "1_1" in row.fname: continue
#     print_limits(row, scale='SM')

# print "\n"
# for index,row in df_results.iterrows():
#     if not "K6" in row.fname: continue
#     if not "1_1" in row.fname: continue
#     print_limits(row, scale='SM')

for index,row in df_results.iterrows():
    if not "K5b" in row.fname: continue
    if not "1_m1" in row.fname: continue
    print_limits(row, scale='ITCK5b')

print "\n"

for index,row in df_results.iterrows():
    if not "K5b" in row.fname: continue
    if not "1_m0" in row.fname: continue
    print_limits(row, scale='KT0K5b')

print "\n"

for index,row in df_results.iterrows():
    if not "K5b" in row.fname: continue
    if not "1_1" in row.fname: continue
    print_limits(row, scale='SMtH')
```
